```python
# Import required libraries
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the cryptocurrency market data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",  # Path to the CSV file
    index_col="coin_id"                 # Set the 'coin_id' column as the index
)

# Display the first 10 rows of the DataFrame
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot a line chart of the data to visualize trends
df_market_data.hvplot.line(
    width=800,  # Set plot width
    height=400, # Set plot height
    rot=90      # Rotate x-axis labels by 90 degrees
)
```

---

### Prepare the Data

```python
# Standardize the data using StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_market_data)  # Normalize numerical values
```

```python
# Convert the scaled data back into a Pandas DataFrame
df_market_data_scaled = pd.DataFrame(
    scaled_data,                       # Use scaled data
    columns=df_market_data.columns     # Set original column names
)

# Add the original 'coin_id' index back to the DataFrame
df_market_data_scaled["coin_id"] = df_market_data.index

# Set 'coin_id' as the index of the new DataFrame
df_market_data_scaled = df_market_data_scaled.set_index("coin_id")

# Display the first five rows of the scaled DataFrame
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12)) 
```

```python
# Initialize an empty list to store inertia values
inertia = []

# Compute the inertia for each k-value
for k in k_values:
    # Create a KMeans model with the current k
    k_model = KMeans(n_clusters=k, random_state=0)
    
    # Fit the model to the scaled data
    k_model.fit(df_market_data_scaled)
    
    # Append the inertia value to the list
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with k-values and corresponding inertia values
elbow_data = {"k": k_values, "inertia": inertia}

# Convert the dictionary into a Pandas DataFrame
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot the Elbow Curve to identify the optimal number of clusters
df_elbow.hvplot.line(
    x="k",           # x-axis: k-values
    y="inertia",     # y-axis: inertia values
    title="Elbow Curve",  # Title for the plot
    xticks=k_values      # Set x-axis ticks to match k-values
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 4

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Set the best k-value based on the Elbow Curve
best_k = 4

# Create a KMeans model with the optimal number of clusters
model = KMeans(n_clusters=best_k, random_state=0)
```

```python
# Fit the K-Means model to the scaled data
model.fit(df_market_data_scaled)
```

```python
# Predict the cluster labels for the scaled data
predicted_clusters = model.predict(df_market_data_scaled)

# Print the resulting array of cluster labels
print(predicted_clusters)
```

```python
# Create a copy of the DataFrame
df_market_data_predictions = df_market_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_predictions["predicted_cluster"] = predicted_clusters

# Display sample data
df_market_data_predictions.head()
```

```python
# Create a scatter plot to visualize the clusters based on original data
df_market_data_predictions.hvplot.scatter(
    x="price_change_percentage_24h",  # x-axis: 24-hour price change
    y="price_change_percentage_7d",   # y-axis: 7-day price change
    by="predicted_cluster",           # Color points by cluster
    hover_cols=["coin_id"],           # Show 'coin_id' on hover
    title="Cryptocurrency Clusters: 24h vs. 7d Price Change"
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model to reduce the data to three components
pca = PCA(n_components=3)
```

```python
# Fit and transform the scaled data using PCA
pca_data = pca.fit_transform(df_market_data_scaled)

# Create a new DataFrame with the PCA-transformed data
df_market_pca = pd.DataFrame(
    data=pca_data,                   # PCA-transformed data
    columns=["PC1", "PC2", "PC3"],   # Principal component labels
    index=df_market_data.index       # Preserve original index
)

# Display the first five rows of the PCA DataFrame
df_market_pca.head()
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance of each component: {explained_variance}")
print(f"Total explained variance: {explained_variance.sum()}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?
